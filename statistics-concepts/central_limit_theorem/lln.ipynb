```python
import numpy as np
import random

import matplotlib.pyplot as plt
import matplotlib.animation as animation
from celluloid import Camera
import seaborn as sns
```

```python
# Demonstration of the LLN using 

coin = np.array([0,1])

# Simulating a coin toss
n_sim = 10

outcomes = []
probs = []
for _ in range(n_sim):
    outcomes.append(np.random.choice(coin))
    probability = np.mean(outcomes)
    probs.append(probability)

fig, ax = plt.subplots(1,2, figsize=(10,5))

ax[1].plot(probs)
ax[1].set_title('Probability of Heads')
ax[1].hlines(0.5, 0, n_sim, linestyles='dashed')

ax[0].plot(outcomes)
ax[0].set_title('Outcome of Heads')

plt.show()

# n_flips = 10000
# outcome = np.random.choice(coin, size=n_flips)

# prob_1 = np.cumsum(outcome)/np.arange(1,n_flips+1)
# prob_0 = 1 - prob_1
```

```python
possible_outcomes = ["heads", "tails"]

def flip_coin():
    result = random.choice(possible_outcomes)
    return result


probability_values = []

def simulate_coin_flip(simulations):
    results = 0
    for sim in range(1, simulations):
        coin_flip = flip_coin()
        if coin_flip == "heads":
            results += 1

        temp_probability = results / sim
        probability_values.append(temp_probability)

    final_probability = results / simulations


    
    plt.style.use('fivethirtyeight')
    plt.title('Monte Carlo Simulations of Coin Flips', fontsize=16)
    plt.axhline(0.5, color='g')
    plt.xlabel('Number of Coin Flips', fontsize=12)
    plt.ylabel('Probability Values', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.plot(probability_values)
    plt.annotate(f'Estimated Probability Value = {final_probability}',
                 xy=(1, 0),
                 xycoords='axes fraction',
                 xytext=(-20, 150),
                 textcoords='offset pixels',
                 horizontalalignment='right',
                 verticalalignment='top')
    plt.tight_layout()
    # plt.savefig('images/coin_flip_simulation.png', dpi=300)

    # return final_probability
    print(final_probability)


simulate_coin_flip(1000)
```
